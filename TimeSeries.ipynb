```python
import pandas as pd
import numpy as np

%matplotlib inline 
```

```python
df= pd.read_csv(r'TimeSeries/airline_passengers.csv', index_col=0, parse_dates=True)
```

```python
df=df.dropna()
```

```python
df.index.freq = 'MS'
```

```python
df.index
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
Span=12
alpha =2/(Span+1)
```

```python
df['EWMA'] = df['Thousands of Passengers'].ewm(alpha=alpha, adjust=False).mean()
```

```python
model = SimpleExpSmoothing (df['Thousands of Passengers'])
```

```python
fitted_model = model.fit(smoothing_level = alpha, optimized=False)
```

```python
model.fit(smoothing_level = alpha, optimized=False).fittedvalues
```

```python
fitted_model.fittedvalues
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df['DES_add_12'] = ExponentialSmoothing(df['Thousands of Passengers'], trend='mul').fit().fittedvalues.shift(-1)
```

```python
df.plot(figsize= (12,4))
```
